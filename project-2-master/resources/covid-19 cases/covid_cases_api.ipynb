```python
import pandas as pd
import requests
import time
import pprint as pp
```

```python
url = "https://covidtracking.com/api/states/daily"
response = requests.get(url).json()
```

```python
date = []
dateChecked = []
state = []
positive = []
negative = []
pending = []
hospitalizedCurrently = []
hospitalizedCumulative = []
inIcuCurrently = []
inIcuCumulative = []
onVentilatorCurrently = []
onVentilatorCumulative = []
recovered = []
hash = []
dateChecked = []
death = []
hospitalized = []
total = []
totalTestResults = []
posNeg = []
fips = []
deathIncrease = []
hospitalizedIncrease = []
negativeIncrease = []
positiveIncrease = []
totalTestResultsIncrease = []



for row in response:
    try:
        date.append(row['date'])
        state.append(row['state'])
        positive.append(row['positive'])
        hash.append(row['hash'])
        dateChecked.append(row['dateChecked'])
        total.append(row['total'])
        totalTestResults.append(row['totalTestResults'])
        posNeg.append(row['posNeg'])
        fips.append(row['fips'])
        deathIncrease.append(row['deathIncrease'])
        hospitalizedIncrease.append(row['hospitalizedIncrease'])
        negativeIncrease.append(row['negativeIncrease'])
        positiveIncrease.append(row['positiveIncrease'])
        totalTestResultsIncrease.append(row['totalTestResultsIncrease'])
        
        try:
            negative.append(row['negative'])
        except:
            negative.append('0')
        try:
            pending.append(row['pending'])
        except:
            pending.append('0')
        try:
            recovered.append(row['recovered'])
        except:
            recovered.append('0')
        try:
            death.append(row['death'])
        except:
            death.append('0')
        try:
            hospitalizedCurrently.append(row['hospitalizedCurrently'])
        except:
            hospitalizedCurrently.append('0')
        try:
            hospitalizedCumulative.append(row['hospitalizedCumulative'])
        except:
            hospitalizedCumulative.append('0')
        try:
            inIcuCurrently.append(row['inIcuCurrently'])
        except:
            inIcuCurrently.append('0')
        try:
            inIcuCumulative.append(row['inIcuCumulative'])
        except:
            inIcuCumulative.append('0')
        try:
            onVentilatorCurrently.append(row['onVentilatorCurrently'])
        except:
            onVentilatorCurrently.append('0')
        try:
            onVentilatorCumulative.append(row['onVentilatorCumulative'])
        except:
            onVentilatorCumulative.append('0')
        try:
            hospitalized.append(row['hospitalized'])
        except:
            hospitalized.append('0')
            
    except (KeyError, IndexError):
        print("---------------------")
        print(f"There was an issue collecting data on row:")
        print("---------------------")
        print(row)


#pp.pprint(response)
```

```python
covid_cases_dict = {
    "date": date,
    "state": state,
    "positive": positive,
    "negative": negative,
    "pending": pending,
    "hospitalizedCurrently": hospitalizedCurrently,
    "hospitalizedCumulative": hospitalizedCumulative,
    "inIcuCurrently": inIcuCurrently,
    "inIcuCumulative": inIcuCumulative,
    "onVentilatorCurrently": onVentilatorCurrently,
    "onVentilatorCumulative": onVentilatorCumulative,
    "recovered": recovered,
    "hash": hash,
    "dateChecked": dateChecked,
    "death": death,
    "hospitalized": hospitalized,
    "total": total,
    "totalTestResults": totalTestResults,
    "posNeg": posNeg,
    "fips": fips,
    "deathIncrease": deathIncrease,
    "hospitalizedIncrease": hospitalizedIncrease,
    "negativeIncrease": negativeIncrease,
    "positiveIncrease": positiveIncrease,
    "totalTestResultsIncrease": totalTestResultsIncrease
}
covid_cases_data = pd.DataFrame(covid_cases_dict)
covid_cases_data
```

```python
covid_cases_data.to_csv(f'covid_cases_{date[0]}.csv')
```
