```python
series_name = []
date = []
value = [] ##dollars per gallon

for api_ids in eia_api_id:
    
    url = 'http://api.eia.gov/series/?api_key=' + eia_api_key + '&series_id=' + api_ids
    response = requests.get(url).json()
    
    series = response['series']
    series_all = series[0]
    series_data = series_all['data']
    
    series_data_name = series_all['name'] ##to append series_name
    
    for row in series_data:
               
        try:
            
            series_data_date = row[0] ##date
            series_data_price = row[1] ##price
            
            series_name.append(series_data_name)
            date.append(series_data_date)
            value.append(series_data_price)

        except (KeyError, IndexError):
            print("---------------------")
            print(f"There was an issue collecting data on row:")
            print("---------------------")
            print(row)


#pp.pprint(response)
```

```python
gas_prices_dict = {
    "series_name": series_name,
    "date": date,
    "dollars_per_gallon": value
}
gas_prices_data = pd.DataFrame(gas_prices_dict)
gas_prices_data
```

```python
gas_prices_data.to_csv(f'gas_prices.csv')
```
